### Problem Definition 

In graph theory, the graph bandwidth problem is to label the $n$ vertices $v_i$ of a graph $G$ with distinct integers so that the quantity 
$\max \{|f(v_i) - f(v_j)| : v_iv_j \in E(G)\}$ is minimized. It is a combinatorial optimization problem and is known to be NP-hard, and it is also known to be NP-hard to approximate, therefore an approximation algorithm does not exist. When the numbering is restricted to the set $\{1, 2, \dots, n\}$, then it is called a perfect numbering.

### Evolutionary Algorithm Implementation
#### Chromosome Representation and Fitness Function
The chromosome is represented as a dictionary which maps (labels) a vertex to an integer number. The population is initialized by assigning random numberings to the different dictionaries.  

The fitness function is the equation $\max \{|f(v_i) - f(v_j)| : v_iv_j \in E(G)\}$, which is implemented by iterating over the edges and computing the maximum value. In the population, the chromosome are stored as a tuple along with their fitness, i.e. (chromosome, fitness(chromosome)).

```python
g_bandwidth.init_population()
print("A single chromosome: \n")
print(g_bandwidth.population[0][0])
print("\nIndividual in Population: \n")
print(g_bandwidth.population[0])
print("\nEntire Population: \n")
print(g_bandwidth.population)
```
